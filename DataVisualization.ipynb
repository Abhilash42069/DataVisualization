```python
df = pd.read_csv("covid_19_data.csv")
```

```python
df1 = pd.read_csv("covid_19_data.csv")
df1=df1.drop(['SNo','Province/State','ObservationDate','Last Update'], axis=1)
df1=df1[14000:16000]
df1.head
```

```python
df=df.drop(['SNo','Country/Region','Province/State','ObservationDate','Last Update'], axis=1)
cpy=df[14000:16000]
cpy=cpy[cpy['Deaths'] >500]
cpy=cpy[cpy['Recovered'] >500]
cpy=cpy.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


ax = plt.axes()
ax=sns.heatmap(cpy)
plt.ylabel('Country/Region', fontsize = 15)
ax.set_title('Covid')
plt.show()
```

```python
contigency= pd.crosstab(df1['Recovered'], df1['Deaths'])
contigency
```

```python
from scipy.stats import chi2_contingency
ctab=pd.crosstab(df1['Recovered'],df1['Deaths'])
ctab
stat, p, dof, expected = chi2_contingency(ctab)
print(p)
alpha = 0.5
if p <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
```
